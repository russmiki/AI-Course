```python
#face detection using opencv
#students : Niloufar zohdy tale / Aylin emamizade / Fatemeh salmani amand / Mani vazirpour / Mohammad heidary /
#description : real-time face detection using haar cascade classifier  

import cv2 #opencv library 
import pyautogui as pg

model = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')  #load haar cascade model
cap =cv2.VideoCapture(0)  #open webcam 

while True:
    x_m , y_m = pg.position() 
    ret , img = cap.read()

    if not ret:
        break

    img = cv2.flip(img , 1)
    cv2.rectangle(img , (210,140) , (400 ,330 ) , (255,255,255) ,5)

    img_gary = cv2.cvtColor(img , cv2.COLOR_BGR2BGRA)  #convert to grayscale 
    face = model.detectMultiScale(img_gary)  #detect faces

    for x,y,a,b in face:
        cv2.rectangle(img , (x,y) , (x+a, y+b) , (0,255,0) ,5) #draw face box

        if x<210:
            pg.moveTo(x_m-10 ,y_m)
        elif x+a>400:
            pg.moveTo(x_m+10 ,y_m)       
        elif y<140:
            pg.moveTo(x_m ,y_m-10)
        elif y+b>330:
            pg.moveTo(x_m ,y_m+10)    

    cv2.imshow('face' , img)
    if cv2.waitKey(1) & 0xFF == 27 :   #exit 
        cv2.destroyAllWindows()
        cap.release()
        break
```
